```python
mail_list=[]
for i in raw_mails:
    mail = i.strip().split()
    refined_mail=[]
    for j in mail:
        if j not in stopwords:
            refined_mail.append(j)
    mail_list.append(refined_mail)
mail_list
```

```python
from collections import defaultdict
d=defaultdict(int)
d['a']=d['a']+1
d
```

```python
d=defaultdict(int)

for i in mail_list:
    for j in i:
        d[j] = d[j] + 1
d
```

```python
refined_data=[]
for i in mail_list:
    temp_mail=[]
    for j in i:
        if d[j]>1:
            temp_mail.append(j)
    refined_data.append(temp_mail)
refined_data
```

```python
#FEATURE EXTRACTION
#bag of words(vocabulary)
import gensim
d2=gensim.corpora.Dictionary(refined_data)
refined_data[0]
d2.doc2bow(refined_data[1])
```

```python
bow_list=[]
for i in refined_data:
    bow_list.append(d2.doc2bow(i))
bow_list
```

```python
count=len(d2)
count
```

```python
#converts into proper matrix
dense_bow=gensim.matutils.corpus2dense(bow_list,num_terms=count).transpose()
dense_bow
```

```python
tags
target=[]
for i in tags:
    if i=='spam':
        target.append(0)
    if i=='ham':
        target.append(1)
target
```

```python
tfidf=gensim.models.TfidfModel(bow_list)
tfidf
```

```python
tfidf_list=tfidf[bow_list]
tfidf
```

```python
#matrix with importance of dictionaries in BOW is formed(by calculating then tfidf formula)
dense_tfidf=gensim.matutils.corpus2dense(tfidf_list,num_terms=count).transpose()
dense_tfidf
```

```python
import sklearn
from sklearn.model_selection import KFold
kf=KFold(n_splits=10,shuffle=True)
dataset=dense_tfidf
sd=list(kf.split(dataset))
sd
```

```python
for i,j in sd[0:1]:
    print(i)
    print(j)
```

```python
import numpy as np
from sklearn import svm
acc_list=[]
for train,test in sd:
    #print(train,test)
    train_data=[]
    train_label=[]
    test_data=[]
    test_label=[]
    for i in train:
        train_data.append(dataset[i])
        train_label.append(target[i])
    for j in test:
        test_data.append(dataset[j])
        test_label.append(target[j])
#train_data,train_label,test_data,test_label
    classifier=svm.SVC()
    classifier.fit(train_data,train_label)
    predict1= classifier.predict(test_data)
    incorrect=(predict1!=test_label).sum() 
    acc=(len(test_label)-incorrect)/len(test_label)
    acc_list.append(acc)
np.array(acc_list).max()
```
